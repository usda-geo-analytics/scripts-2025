```python
import os
import pandas as pd
```

```python
dir = r'C:\Users\misti.wudtke\OneDrive - USDA\PROJECTS\FY25_Q2_EmployeeLocations_50mi_MatthGoldsboro\Input Files'

file_list = [os.path.join(dir, f) for f in os.listdir(dir) if f.split('.')[-1] in ['xlsx', 'csv']]
```

```python
dfs = []

agencies = ['NRCS', 'FSA', 'RMA']

convert_cols = ['Site ID', 'Office ID', 'Total Staff', 'Total Seats', 'Available Seats']

for fl in file_list:

    if fl.endswith('.xlsx'):
        with pd.ExcelFile(fl) as fobj:
            for sheet in fobj.sheet_names:
                df = pd.read_excel(fobj, sheet)

    elif fl.endswith('.csv'):
        df = pd.read_csv(fl)

    for c in convert_cols:
        if c in df.columns:
            df[c] = df[c].astype(float)

    for a in agencies:
        if a in fl:
            df['Agency'] = a

    dfs.append(df)
```

```python
smasht = pd.concat(dfs)

smasht.shape
```

```python
smasht.to_excel('First_Output.xlsx', index=False)

smasht.head()
```

```python
cols = ['Site ID',
        'Site Name',
        'Office ID',
        'Office Name',
        'Agency',
        'State Name',
        'Address',
        'City',
        'State Abbr',
        'Zip',
        'Total Staff',
        'Total Seats',
        'Available Seats']

distil1 = smasht[cols]

distil1.head()
```

```python
# Here is a handy block for removing all leading/trailing whitespace
# in all string columns in the df:
for col in distil1.columns:
    if distil1[col].dtype == 'object':
        distil1[col] = distil1[col].str.strip()
```

```python
distil1.to_excel('Second_Output_Distilled.xlsx', index=False)

distil1.shape
```

```python
dups = distil1[distil1['Site ID'].duplicated()]

dups.to_excel('dups.xlsx', index=False)

dups.shape
```

```python
all_dups = distil1[distil1['Site ID'].isin(dups['Site ID'])]

all_dups.to_excel('all_dups.xlsx', index=False)

all_dups.shape
```
